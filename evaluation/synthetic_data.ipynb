## Synthetic data generation, testing and visualization
Script to test and generate samples from the developed synthetic data class to assess how it works, the correctness, etc etc etc


```python
#Import
# working dir
%cd /homedtic/gmarti/CODE/RNN-VAE/

# Imports
import sys
import os
sys.path.insert(0, '/homedtic/gmarti/CODE/RNN-VAE/')
import math
from rnnvae.data_gen import SinDataGenerator
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Curves 
curves = [
    ("sigmoid", {"L": -3, "k": 3, "x0": 2}),    
    ("sigmoid", {"L": 2, "k": -1, "x0": 5}),
    ("sigmoid", {"L": -2, "k": 2, "x0": 7})
    ]

ntp = 15
noise = 0.5
variable_tp=False
equal_spacing=True
```

```python
#Create class object
gen_model = SinDataGenerator(curves, ntp, noise, variable_tp)
```

```python
feat = 0
#Generate a random sample and visualize it with the base signal
x, y = gen_model.generate_full_signal()
print(x.shape)
print(y.shape)
# X, Y = gen_model.sample()
plt.plot(x, y[:, feat], '-', c='blue')
# plt.scatter(x=X, y=Y[feat, :], color='black', marker='x')
plt.show()
```

```python
#Generate n random samples
x, y = gen_model.generate_full_signal()
plt.plot(x, y[:, feat], '-', c='blue')
samples = gen_model.generate_n_samples(500)
for (X, Y) in samples:
    # plt.scatter(x=list(range(len(sample[0]))), y=sample[1], color='black', marker='x')
    plt.scatter(x=X, y=Y[:, feat], color='black', marker='x')
plt.show()
```

```python
# create subplots
sns.set()
fig, ((ax1, ax2, ax3)) = plt.subplots(1, 3, figsize=(10,3), sharex=False, sharey=False)
## Create figures of three signals and their points, with a subplots and everything
feats = [0,1,2]
axes = [ax1,ax2,ax3]
for (f, ax) in zip(feats, axes):
    x, y = gen_model.generate_full_signal()
    ax.plot(x, y[:, f], '-', c='blue')
    samples = gen_model.generate_n_samples(500)
    for (X, Y) in samples:
        # plt.scatter(x=list(range(len(sample[0]))), y=sample[1], color='black', marker='x')
        ax.scatter(x=X, y=Y[:, f], color='black', marker='x', s=5)
# remove the x and y ticks
for ax in axes:
    #ax.set_xticks([])
    #ax.set_yticks([])
    ax.set_xticklabels([])
    ax.set_yticklabels([])
fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
#plt.savefig("samples_long_2.pdf", dpi=300)
plt.show()
```
